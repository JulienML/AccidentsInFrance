We could affirm purely based on the correlation matrix that the severity of the accident may be directly correlated linearly to the user category, maybe to the sex of the individual, the year of birth of the person and the security precautions applied (seatbelt, etc.)
Surprisingly, we don't see a direct linear correlation between the weather or the luminosity with the severity of the car accident.
Let's apply the PCA to have a more detailed analysis.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Circle
 
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.metrics.pairwise import euclidean_distances

data = data.fillna(data.mean())
X = data.to_numpy()
Xstd = StandardScaler().fit_transform(X)
pca = PCA(n_components=2)
Xpca = pca.fit_transform(Xstd)
```

```python
ccircle = []
eucl_dist = []
for i,j in enumerate(X .T):
    corr1 = np.corrcoef(j,Xpca[:,0])[0,1]
    corr2 = np.corrcoef(j,Xpca[:,1])[0,1]
    ccircle.append((corr1, corr2))
    eucl_dist.append(np.sqrt(corr1**2 + corr2**2))
```

```python
with plt.style.context(('seaborn-whitegrid')):
    fig, axs = plt.subplots(figsize=(6, 6))
    for i,j in enumerate(eucl_dist):
        arrow_col = plt.cm.cividis((eucl_dist[i] - np.array(eucl_dist).min())/\
                                (np.array(eucl_dist).max() - np.array(eucl_dist).min()) )
        axs.arrow(0,0, # Arrows start at the origin
                 ccircle[i][0],  #0 for PC1
                 ccircle[i][1],  #1 for PC2
                 lw = 2, # line width
                 length_includes_head=True, 
                 color = arrow_col,
                 fc = arrow_col,
                 head_width=0.05,
                 head_length=0.05)
        axs.text(ccircle[i][0]/2,ccircle[i][1]/2, data.columns[i])
    # Draw the unit circle, for clarity
    circle = Circle((0, 0), 1, facecolor='none', edgecolor='k', linewidth=1, alpha=0.5)
    axs.add_patch(circle)
    axs.set_xlabel("PCA 1")
    axs.set_ylabel("PCA 2")
plt.tight_layout()
plt.show()
```

Here, we can affirm many point : 
First of all, there are many aspects of the data thare are represented here. Indeed, the data is represented on differents independant axis, for example road surface and User category. Regarding the orientation of the arrows we can affirm the positive or negative correlation between data which are correlated with each other.
For example, we could affirm that the sex of the driver and the year of birth could be positively correlated to the Trip purpose.
